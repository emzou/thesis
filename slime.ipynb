```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import datetime 
import re, os
from pathlib import Path 
import glob
```

```python
data_path = Path.cwd() / "data"
tsv_files = []

for file in glob.glob("{}/*.tsv".format(data_path.as_posix())):
    _dg = pd.read_csv(file,sep='\t')
    tsv_files.append(_dg)

merged_octo30 = pd.concat(tsv_files)
```

```python
df = merged_octo30.drop(['SourceId', 'TargetId'], axis = 1)
```

```python
df = df [ df['count']>6]
```

```python
da = df.groupby(['pairs', 'Source', 'Target', 'SourceFreq', 'TargetFreq', 'freqsum']).sum()
dc = pd.DataFrame (da).reset_index()
dc
```

```python
opps = dict(df['pairs'].value_counts())
def oppen (tuple):
    return opps[tuple]
```

```python
dc = pd.DataFrame (da).reset_index()
```

```python
dc['Identities'] = dc['pairs'].apply(oppen)
```

```python
dc
```

```python
dac = list(zip(df['Source'], df['count'].apply(str)))
```

```python
import collections
c = collections.defaultdict(list)
for a,b in dac: 
    c[a].extend(b)

list2 = list(c.items())
```

```python
def silly (tuple):
    result = [int(i) for i in tuple[1]]
    return tuple[0], sum(result)

list3= [silly(m) for m in list2]
```

```python
dot = pd.DataFrame(list3, columns = ['term', 'count'] )
dot
```

```python
def getter (n):
    return round(n/ dot['count'].max(), 2)
```

```python
dot ['zipf'] = dot['count'].apply(getter)
```

```python
dot
```

```python
zcount = dict(zip(dot['term'], dot['count']))
zfreq = dict (zip(dot['term'], dot['zipf']))
```

```python
def counter (source):
    return zcount[source]

def freqqer (source):
    return zfreq [source]
```

```python
dc ['SourceZfreq'] = dc['Source'].apply(freqqer)
dc ['TargetZfreq'] = dc['Target'].apply(freqqer)
dc ['SourceZCount'] = dc['Source'].apply(counter)
dc ['TargetZCount'] = dc['Target'].apply(counter)
```
