```python
import pandas as pd 
import re
```

```python
df = pd.read_csv('/Users/emilyzou/Desktop/carrot/hasanvideos.csv')
```

```python
da = df.drop(columns = ['channelId', 'channelTitle',  'publishedAt', 'videoCategoryId', 'caption','defaultLanguage', 'defaultLAudioLanguage', 'thumbnail_maxres', 'licensedContent', 'dislikeCount', 'favoriteCount', 'duration', 'durationSec', 'dimension', 'definition'])
```

```python
da ['date'] = da ['publishedAtSQL'].astype('string')
da1 = da[da['date'].str.contains('2022-11 | 2022-12 | 2023-01 |2023-02')]
da1 ['tags'] = da1 ['tags'].astype('string')
da1['tags'] = da1['tags'].apply(lambda x: x.split(','))
```

```python
dc = da1.reset_index()
```

```python
taglist = dc['tags']
```

```python
import itertools
def devilfruit(x):
        zoro = list (itertools.product(x, repeat = 2))
        return zoro

sanji = [devilfruit(x) for x in taglist]

def nami(m): 
    return [i for r in m for i in r ]

nami (sanji)
usopp = nami(sanji)
```

```python
from collections import Counter
c = Counter (zip(usopp))
c
```

```python
brook = pd.DataFrame.from_dict(c, orient = 'index').reset_index()
brook.dtypes
```

```python
brook.columns = ['pairs', 'count']
```

```python
brook
```

```python
brook ['pairs'] = brook ['pairs'].astype('string')
```

```python
brook
```

```python
def robin(s): 
    return s[1:-2]
```

```python
brook ['pairs'] = brook ['pairs'].apply(robin)
```

```python
brook
```

```python
def franky (s):
    return s.split(',')[0]

def chopper (s): 
    return s.split(',')[1]
```

```python
brook ['Source'] = brook ['pairs'].apply(franky)
brook ['Target'] = brook ['pairs'].apply(chopper)
```

```python
brook
```

```python
def rubber(s): 
    return s[2:-1]
def man (s): 
    return s[2:-2]
```

```python
brook ['Source'] = brook ['Source'].apply(rubber)
brook ['Target'] = brook ['Target'].apply(man)
```
